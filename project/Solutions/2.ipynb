```python
import numpy as np
import matplotlib.pyplot as plt
```

## Q1: 

```python
def objective_function(x, c):
    return .5*((x-c)**2) + max(x, 0)

def gradient(x, c):
    return (x-c) + 1*(x>=0)
```

```python
objective_function(-1, 5), gradient(-1, 5)
```

```python
def simulate(step_size, iters:int=20, seed=4):
    np.random.seed(seed)
    #x_0, c = np.random.randint(0, 10000, (2,))
    x_0, c = np.random.randn(2)

    print (f"initial x and c: {x_0, c}")
    x_old = x_0
    x, obj = [x_old, ], [objective_function(x_old, c), ]
    for i in range(iters):
        x_new = x_old - step_size*gradient(x_old, c)
        x_old = x_new
        obj.append(objective_function(x_old, c))
        x.append(x_old)
    return x, obj
```

```python
def plot(x, obj, filename=None):
    fig, axs = plt.subplots(1, 2, figsize=(10,5))
    axs[0].plot(range(len(obj)), obj)
    axs[0].set_title("Objective Function", fontsize=15)
    axs[0].set_xlabel("iter", fontsize=15)
    axs[0].tick_params(axis="both",  labelsize=12)

    axs[1].plot(range(len(x)), x)
    axs[1].set_title("X value", fontsize=15)
    axs[1].set_xlabel("iter", fontsize=15)
    axs[1].tick_params(axis="both",  labelsize=12)
    if filename: fig.savefig(f"{filename}", bbox_inches="tight", pad_inches=0.5)
    plt.show()
```

```python
x, obj = simulate(1, 50, 10)
plot(x, obj, "s=1.png")
#list(zip(x, obj))
```

```python
x, obj = simulate(0.1, 50, 4)
plot(x, obj, "s=0.1.png")
#list(zip(x, obj))
```

```python
x, obj = simulate(0.01, 100, 4)
plot(x, obj, "s=0.01.png")
```
